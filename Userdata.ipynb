```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#load data
data=pd.read_csv('/content/Excelerate userdata.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
#checking missing values
data.isnull().sum()
```

```python
# 1. Missing Values Analysis
missing_values = data.isnull().sum()
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing_report = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
```

```python
#replacing missing values in the degree column
data['Degree'] = data['Degree'].fillna('Not Specified')
```

```python
#cheking missing values
data.isnull().sum()
```

```python
data['Country'].value_counts()
```

```python
data['Country']= data['Country'].fillna('India')
```

```python
#plot a bar plot for the Gender Variable
data['Gender'].value_counts().plot(kind='bar')
colour = sns.color_palette('pastel')[0]
tittle = plt.title('Gender Count')
x = plt.xlabel('Gender')
y = plt.ylabel('Count')
```

```python
pie = data['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
tittle = plt.title('Gender Count')
```

```python
# Country-wise Distribution (Top 10 countries)
country_distribution = data['Country'].value_counts().head(10)
country_distribution
```

```python
#Sign-up Date Trends (Convert to datetime and extract yearly trends)
data['Sign Up Date'] = pd.to_datetime(data['Sign Up Date'], errors='coerce')
signup_trends = data['Sign Up Date'].dt.year.value_counts().sort_index()
signup_trends
```

```python
#plot of gender and sign up date
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Gender', hue='isFromSocialMedia')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Degree', hue='isFromSocialMedia')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
```
